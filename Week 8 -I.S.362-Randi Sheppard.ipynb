```python
import pandas as pd
import numpy as np

column_names = [
    'mpg', 'cylinders', 'displacement', 'horsepower',
    'weight', 'acceleration', 'model_year', 'origin', 'car_name'
]

df = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data',
    sep='\s+',
    names=column_names,
    na_values='?'
)

```

```python
print(df['origin'].unique())
print(df['origin'].value_counts(dropna=False))
```

```python
df['origin'] = df['origin'].astype(int)
origin_map = {1: 'USA', 2: 'Asia', 3: 'Europe'}
df['origin'] = df['origin'].map(origin_map)
```

```python
print(df['origin'].unique())
```

```python
print(df[['origin', 'car_name']].head(10))
```

```python
# Step 1: Convert origin to numeric (if it's still strings like '1', '2', '3')
df['origin'] = pd.to_numeric(df['origin'], errors='coerce')

# Step 2: Map numeric codes to region names
origin_map = {1: 'USA', 2: 'Asia', 3: 'Europe'}
df['origin'] = df['origin'].map(origin_map)
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
origin_map = {1: 'USA', 2: 'Asia', 3: 'Europe'}
df['origin'] = df['origin'].map(origin_map)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='cylinders', data=df, palette='Set2')
plt.title('Distribution of Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
##This bar chart shows the frequency of different cylinder counts Most cars in the dataset have 4 or 8 cylinders, with fewer cars having 3, 5, or 6.##
```

```python
sns.scatterplot(x='horsepower', y='weight', data=df, hue='origin', palette='Set1')
plt.title('Horsepower vs Weight by Origin')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.show()
```

```python
##This scatterplot reveals a positive correlation between horsepower and weight.
Heavier cars tend to have more powerful engines. The color hue shows regional differences in vehicle design##
```

```python
##Question: Do cars with more cylinders tend to have lower MPG?##
sns.boxplot(x='cylinders', y='mpg', data=df, palette='coolwarm')
plt.title('MPG Distribution by Cylinder Count')
plt.xlabel('Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
Question: Do cars with more cylinders tend to have lower MPG?
```

```python
##The boxplot shows that cars with more cylinders generally have lower miles per gallon (MPG).
4-cylinder cars have the highest MPG on average.
8-cylinder cars show the lowest MPG, confirming that larger engines are less fuel-efficient.##
```
